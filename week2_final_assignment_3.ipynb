```python
%%writefile app.py

import streamlit as st
import pandas as pd
import numpy as np
import plotly.express as px

st.set_page_config(page_title="Smart Analytics Tool", layout="wide")

st.title("📊 Smart Analytics Tool")
st.write("Upload Life Expectancy Dataset")

uploaded_file = st.file_uploader(
    "Upload CSV File",
    type=["csv"]
)

if uploaded_file is not None:

    df = pd.read_csv(uploaded_file)

    # Dataset Preview
    st.subheader("Dataset Preview")
    st.dataframe(df)

    # Dataset Shape
    st.subheader("Dataset Information")
    st.write("Rows:", df.shape[0])
    st.write("Columns:", df.shape[1])

    # Missing Values
    st.subheader("Missing Value Analysis")

    missing = df.isnull().sum()

    st.dataframe(missing)

    fig_missing = px.bar(
        x=missing.index,
        y=missing.values,
        labels={"x":"Columns","y":"Missing Values"},
        title="Missing Values"
    )

    st.plotly_chart(fig_missing)

    # Statistical Summary
    st.subheader("Statistical Summary")

    st.dataframe(df.describe())

    # Numerical Columns
    numeric_cols = df.select_dtypes(include=np.number).columns

    # Visualization 1
    st.subheader("Visualization 1: Histogram")

    hist_col = st.selectbox(
        "Select Column",
        numeric_cols
    )

    fig1 = px.histogram(
        df,
        x=hist_col,
        title=f"Distribution of {hist_col}"
    )

    st.plotly_chart(fig1)

    # Visualization 2
    st.subheader("Visualization 2: Scatter Plot")

    x_col = st.selectbox(
        "X Axis",
        numeric_cols,
        key="x"
    )

    y_col = st.selectbox(
        "Y Axis",
        numeric_cols,
        key="y"
    )

    fig2 = px.scatter(
        df,
        x=x_col,
        y=y_col,
        title=f"{x_col} vs {y_col}"
    )

    st.plotly_chart(fig2)

    # Visualization 3
    st.subheader("Visualization 3: Line Chart")

    line_col = st.selectbox(
        "Select Column for Trend",
        numeric_cols,
        key="line"
    )

    fig3 = px.line(
        df,
        y=line_col,
        title=f"Trend of {line_col}"
    )

    st.plotly_chart(fig3)
```
